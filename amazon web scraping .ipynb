```python
# import libraries 

from bs4 import BeautifulSoup
import requests 
import smtplib
import time
import datetime 
```

```python
URL = 'https://www.amazon.fr/-/en/dp/B0F4NXKB76'

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/147.0.0.0 Safari/537.36",
    "Accept-Language": "en-US, en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "DNT": "1",
    "Connection": "close",
    "Upgrade-Insecure-Requests": "1"
}

page = requests.get(URL, headers=headers)
soup1 = BeautifulSoup(page.content, "html.parser")
soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

# TITLE
title = soup2.find(id='productTitle').get_text().strip()

# PRICE - try multiple methods in order of reliability
price = None

# Method 1: a-price-whole + a-price-fraction (most reliable right now)
whole = soup2.find('span', class_='a-price-whole')
fraction = soup2.find('span', class_='a-price-fraction')

if whole and fraction:
    whole_text = whole.get_text().strip().replace(".", "").replace(",", "").replace("\n", "").replace(" ", "")
    fraction_text = fraction.get_text().strip().replace("\n", "").replace(" ", "")
    price = whole_text + "." + fraction_text
    
# Method 2: Fallback to a-offscreen
if not price:
    offscreen = soup2.find('span', class_='a-offscreen')
    if offscreen:
        price = offscreen.get_text().strip()

# Method 3: Still nothing? Tell us
if not price:
    price = "PRICE NOT FOUND"

print("Title:", title)
print("Price:", price)
```

```python
def check_price():
    # Step 1: URL
    URL = 'https://www.amazon.fr/-/en/dp/B0F4NXKB76'
    
    # Step 2: Headers
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/147.0.0.0 Safari/537.36",
        "Accept-Language": "en-US, en;q=0.9",
        "Accept-Encoding": "gzip, deflate, br",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
        "DNT": "1",
        "Connection": "close",
        "Upgrade-Insecure-Requests": "1"
    }
    
    # Step 3: Get page
    page = requests.get(URL, headers=headers)
    soup1 = BeautifulSoup(page.content, "html.parser")
    soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
    
    # Step 4: Get title
    title = soup2.find(id='productTitle').get_text().strip()
    
    # Step 5: Get price (using our fallback method)
    whole = soup2.find('span', class_='a-price-whole')
    fraction = soup2.find('span', class_='a-price-fraction')
    
    if whole and fraction:
        whole_text = whole.get_text().strip().replace(".", "").replace(",", "").replace("\n", "").replace(" ", "")
        fraction_text = fraction.get_text().strip().replace("\n", "").replace(" ", "")
        price = whole_text + "." + fraction_text
    else:
        price = "Price not found"
    
    # Step 6: Today's date
    today = datetime.date.today()
    
    # Step 7: Append to CSV
    data = [title, price, today]
    
    with open('AWSDataset.csv', 'a+', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(data)

# ⭐ EMAIL ALERT LOGIC
    if float(price) < 20:
        print("🎉 Price dropped below 20! Sending email...")
        send_mail()
    else:
        print("Price still high. No email.")
    
```
